```python
# Definir el orden correcto de los rangos
orden_rangos = [
    'Hasta 1 SMLMV',
    'Mayor a 1 SMLMV y menor o igual a 12 SMLMV',
    'Mayor a 12 SMLMV y menor o igual a 25 SMLMV',
    'Mayor a 25 SMLMV y menor o igual a 50 SMLMV',
    'Mayor a 50 SMLMV y menor o igual a 100 SMLMV',
    'Mayor a 100 SMLMV y menor o igual a 120 SMLMV',
    'Mayor a 120 SMLMV y menor o igual a 240 SMLMV',
    'Mayor a 240 SMLMV y menor o igual a 480 SMLMV',
    'Mayor a 480 SMLMV'
]

# Convertir a categoría ordenada
df['rango_monto_desembolsado'] = pd.Categorical(
    df['rango_monto_desembolsado'], 
    categories=orden_rangos, 
    ordered=True
)

df['rango_monto_desembolsado']
```

```python
# Ejemplo de cómo comparar DataFrames
df_a = pd.DataFrame({'col1': [1, 2], 'col2': ['a', 'b']})
df_b = df_a.copy()

# 1. Verificar igualdad estricta (True/False)
# Verifica índices, columnas, tipos de datos y valores. Maneja NaNs como iguales.
son_iguales = df_a.equals(df_b)
print(f"¿Son estrictamente iguales? {son_iguales}")

# 2. Comparación elemento por elemento
# Retorna un DataFrame de booleanos. Cuidado: NaNs != NaNs aquí.
diferencias = df_a == df_b
print("\nComparación elemento a elemento:\n", diferencias)

# 3. Herramienta de testing (muy útil para validar ETLs)
# Lanza una excepción si hay diferencias. Ofrece detalles precisos.
try:
    pd.testing.assert_frame_equal(df_a, df_b)
    print("\nLa prueba de aserción pasó correctamente.")
except AssertionError as e:
    print(f"\nError en la comparación: {e}")
```
